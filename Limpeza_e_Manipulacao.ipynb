```python
dados_limpos = df.loc [:,  ['id', 'listing_url', 'scrape_id', 'last_scraped', 'name', 'description',
                            'host_since', 'latitude', 'longitude', 'room_type', 'accommodates', 'bathrooms',
                            'bedrooms', 'beds', 'price', 'review_scores_rating', 'review_scores_accuracy']]
dados_limpos.info()
```

Há 4 tipos de imóveis:
['Entire home/apt', 'Private room', 'Shared room', 'Hotel room']
<br></br>
E em price, vejo 3 problemas:
cifrão, casas decimais separadas por vírgula e a tipagem como objeto(texto)
'$2,697.00', dtype=object
<br></br>
Já em banheiros e quartos, os dados estão tipados em float. E por curiosidade lavabos sao considerados como 0.5 no número de banheiros.
Também há valores nulos em muitas variáveis.

```python
#Primeiro limpamos a variável price
dados_limpos['price'] = dados_limpos['price'].str.replace('$', '')
dados_limpos['price'] = dados_limpos['price'].str.replace(',', '')
dados_limpos['price'].unique()
```

```python
#Agora convertemos para a tipagem correta
dados_limpos['price'] = dados_limpos['price'].astype(float)
dados_limpos['price'].unique()
```

```python
dados_limpos.info()
```

```python
(dados_limpos.isnull().sum() / len(dados_limpos)) * 100
```

Quando os dados nulos forem menores que 10% da base podem ser apagados, mas apra apagar deve haver um motivo. Para análise descritiva e estatística básica não há necessidade de apagar, mas para ML e clusterização geralmente as bibliotecas já vem prontas para lidar com os nulos apagando/ignorando ou será preciso apagá-los manualmente.

A limpeza dos nulos será feita apenas em price, bathrooms, bedrooms e beds. Já que: reviews, host_since e description não serão relevantes ter nulos ou não.

```python
dados_limpos = dados_limpos.dropna(subset=['price', 'bathrooms', 'bedrooms', 'beds'])
dados_limpos.info()
```

```python
#Salvar csv para posteridade
dados_limpos.to_csv("dados_limpos.csv", index=False)
```
