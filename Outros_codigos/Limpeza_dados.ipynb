# Início da limpeza dos dados


  - Começo selecionando apenas as colunas de interesse

  - Após isso resolver os valores e tipagem na coluna 'price'

  - E em price, vejo 3 problemas: cifrão, casas decimais separadas por vírgula e a tipagem como objeto(texto) '$2,697.00', dtype=object

  - Também vou padronizar o conteúdo das variáveis de texto

  - Farei a conversão nas variáveis: host_since e last_scraped para um formato de data válido

  - Já em banheiros e quartos, os dados estão tipados em float. E por curiosidade lavabos sao considerados como 0.5 no número de banheiros.
  
  - Também há valores nulos em muitas variáveis, esses valores serão tratados pois para o teste das hipóteses, os dados precisam ser válidos.

  - Os dados das colunas latitude e longitude estão salvos como texto (object) e possuem vírgula como separador, então farei basicamente o mesmo procedimento que fiz em price. Tirando a vírgula e passando o formato em float para serem acessados pelo cluster posteriormente.

```python
#Listando as colunas para facilitar a copiar as colunas que quero
lista_de_colunas = df.columns
print(lista_de_colunas)
```

```python
dados_limpos = df.loc [:,  ['id', 'listing_url', 'scrape_id', 'last_scraped', 'name', 'description',
                            'host_since', 'neighbourhood_cleansed', 'latitude', 'longitude', 'room_type', 'accommodates', 'bathrooms',
                            'bedrooms', 'beds', 'price', 'review_scores_rating', 'review_scores_accuracy']]
dados_limpos.info()
```

```python
#Loop para a quantidade e tipos de dados da base
for coluna in dados_limpos.columns:
  print("==> ", coluna, " QTD únicos: ", dados_limpos[coluna].nunique())
  print(dados_limpos[coluna].unique())
  print()
```

```python
#Primeiro limpamos a variável price
dados_limpos['price'] = dados_limpos['price'].str.replace('$', '')
dados_limpos['price'] = dados_limpos['price'].str.replace(',', '')
dados_limpos['price'].unique()
```

```python
#Agora converto para a tipagem correta
dados_limpos['price'] = dados_limpos['price'].astype(float)
dados_limpos['price'].unique()
```

```python
dados_limpos['name'] = dados_limpos['name'].str.upper()
dados_limpos['description'] = dados_limpos['description'].str.upper()
dados_limpos['neighbourhood_cleansed'] = dados_limpos['neighbourhood_cleansed'].str.upper()
dados_limpos['name'].unique()
```

```python
dados_limpos.host_since = pd.to_datetime(dados_limpos.host_since, format='%Y-%m-%d')
```

```python
dados_limpos.last_scraped = pd.to_datetime(dados_limpos.last_scraped, format='%Y-%m-%d')
```

```python
dados_limpos = dados_limpos.dropna(subset=['price', 'bathrooms', 'bedrooms', 'beds'])
dados_limpos.info()
```

```python
dados_limpos['latitude'] = dados_limpos['latitude'].astype(str).str.replace(',', '.')
dados_limpos['longitude'] = dados_limpos['longitude'].astype(str).str.replace(',', '.')

dados_limpos['latitude'] = dados_limpos['latitude'].astype(float)
dados_limpos['longitude'] = dados_limpos['longitude'].astype(float)
```
